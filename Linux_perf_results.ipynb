```python
import matplotlib.pyplot as plt
import numpy as np

# Data from provided information
matrix_types = ['Tiled MM', 'Loop Interchange MM', 'Normal MM']

# Cache Misses data
cache_misses = [11515323, 20592675, 26329638]
cache_references = [195532980, 512181064, 4750344499]

# L1 Cache Misses data
l1_cache_load_misses = [474209970, 221882788, 5423315507]
l1_cache_loads = [155500085170, 151931891770, 151936528887]

# LLC Cache Misses data
llc_load_misses = [754471, 233108, 7509692]
llc_loads = [26123566, 18644540, 1591546892]

# Plotting Cache Misses vs. Cache References
plt.figure(figsize=(12, 6))
bar_width = 0.3
index = np.arange(len(matrix_types))

plt.bar(index, cache_misses, bar_width, label='Cache Misses', alpha=0.7)
plt.bar(index + bar_width, cache_references, bar_width, label='Cache References', alpha=0.7)

plt.xlabel('Matrix Multiplication Type')
plt.ylabel('Count')
plt.title('Cache Misses and Cache References Comparison')
plt.xticks(index + bar_width / 2, matrix_types)
plt.legend()

# Adding value labels on top of the bars
for i, v in enumerate(cache_misses):
    plt.text(i - 0.1, v + 5000000, str(v), color='black', fontweight='bold')

for i, v in enumerate(cache_references):
    plt.text(i + bar_width - 0.1, v + 50000000, str(v), color='black', fontweight='bold')

plt.tight_layout()
plt.show()

# Plotting L1 Cache Load Misses vs. L1 Cache Loads
plt.figure(figsize=(12, 6))
plt.bar(index, l1_cache_load_misses, bar_width, label='L1 Cache Load Misses', alpha=0.7)
plt.bar(index + bar_width, l1_cache_loads, bar_width, label='L1 Cache Loads', alpha=0.7)

plt.xlabel('Matrix Multiplication Type')
plt.ylabel('Count')
plt.title('L1 Cache Load Misses and L1 Cache Loads Comparison')
plt.xticks(index + bar_width / 2, matrix_types)
plt.legend()

# Adding value labels on top of the bars
for i, v in enumerate(l1_cache_load_misses):
    plt.text(i - 0.1, v + 50000000, str(v), color='black', fontweight='bold')

for i, v in enumerate(l1_cache_loads):
    plt.text(i + bar_width - 0.1, v + 500000000, str(v), color='black', fontweight='bold')

plt.tight_layout()
plt.show()

# Plotting LLC Load Misses vs. LLC Loads
plt.figure(figsize=(12, 6))
plt.bar(index, llc_load_misses, bar_width, label='LLC Load Misses', alpha=0.7)
plt.bar(index + bar_width, llc_loads, bar_width, label='LLC Loads', alpha=0.7)

plt.xlabel('Matrix Multiplication Type')
plt.ylabel('Count')
plt.title('LLC Load Misses and LLC Loads Comparison')
plt.xticks(index + bar_width / 2, matrix_types)
plt.legend()

# Adding value labels on top of the bars
for i, v in enumerate(llc_load_misses):
    plt.text(i - 0.1, v + 200000, str(v), color='black', fontweight='bold')

for i, v in enumerate(llc_loads):
    plt.text(i + bar_width - 0.1, v + 20000000, str(v), color='black', fontweight='bold')

plt.tight_layout()
plt.show()
```
